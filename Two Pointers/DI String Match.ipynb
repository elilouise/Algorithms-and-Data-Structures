## Approach Overview:

Scan through the characters in `s` (`I` or `D`) and at each iteration determine which integer (between `0 to n`) we should pick and add to `perm`. 

## Detailed Explanation:

This was a fun one. 

We're given a string `s` consisting of a series of characters (either `D` or `I`) that simultaneously instructs us how to populate and re-construct `perm`. 

The rule is this. 

- If `s[i] = "I"` then this means `perm[i] < perm[i + 1]`

- If `s[i] = "D"` then this means `perm[i] > perm[i + 1]`

- `perm[i]` can be any any number between `0 and n` , where `n = length of s`

At any given index (of `perm`) we need to decide which integer to choose and add to it to `perm`. The challenge is which one do we choose. Ofcourse we need to use our rules to guide us. 

If `s[i] = "I"` then this means `perm[i] < perm[i + 1]`:

So for the current index we must pick a value that's LESS than the next successive element in `perm`. To ensure this is always the case, we pick the smallest value possible from the integer range.

*This ensures that the next successive element is always going to be greater than the current one (because we picked the smallest value possible).*

Conversely, if `s[i] = "D"` then this means `perm[i] > perm[i + 1]`

So for the current index we must pick a value that's GREATER than the next successive element in `perm`. To ensure this is always the case, we pick the largest value possible from the integer range. 

*This ensures that the next successive element is always going to be less than the current one (because we picked the largest value possible).*

## Solution:

```python
class Solution:
    def diStringMatch(self, s: str) -> List[int]:
        
        # initialize left and right pointer
        left, right = 0, len(s)
        # initialize `perm` array
        perm = []
        
        # loop through every character in s
        for i in s:

            # check if current character is "I"
            if i == "I": 
                # add 'left' to perm (because it's the smallest)
                # perm[i] < perm[i + 1]
                perm.append(left)
                # increment left 
                left += 1
            else:    
                # add 'right' to perm (because it's the largest)
                perm.append(right)
                # decrement right
                right -= 1

        # add the last remaining integer to perm
        perm.append(left)
        
        # return perm
        return perm
```
