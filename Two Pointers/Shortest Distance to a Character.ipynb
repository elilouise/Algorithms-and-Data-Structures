## Approach Overview (Brute Force):

We loop through `s` to find and store (in an array) all the indices at which `c` occur. We then loop through `s` again, and at each character, we calculate and store (in an output array) the respective smallest distance current character from `c` (using the `c` indices array).

## Detailed Explanation:

Fairly straightforward brute force approach. 

We scan through each character in `s` to locate and store all the positions at which `c` occur. We use an array called `indices` to store the indices.

We then go through each character in `s` and for each we calculate its distance from all the positions of `c` in `s` and store the smallest one in an array called `output`.

We do this for every character in `s`.

# Solution:

```python
class Solution:
    def shortestToChar(self, s: str, c: str) -> List[int]:
        
        # array of all indices of c in s
        indices = []
        # output array of distance answers
        output = []
         
        # loop through every character in s and their corresponding indices
        for index, char in enumerate(s):

            # check if current character equals c
            if char == c:
                # add the index of the character to indices
                indices.append(index)
    
        # loop through every character index in s
        for i in range(len(s)): 

            # initialize distance to len(s) 
            distance = len(s)

            # loop through all the indices in indices
            for index in indices:

                # find the minimum (i to c) distance for current i
                distance = min(distance, abs(i - index))
            
            # add the answer to output array
            output.append(distance)
        
        # return output
        return output

    
    
    
## Time Complexity: O(n^2) 
## Space Complexity: O(n)  
```

## Optimized Solution


## Approach Overview:

Loop through `s` twice. 

The first time to calculate the distance of each character in `s` to the closest occurence of `c` on the **left side**.

And the second time to calculate the distance of each character in `s` to the closest occurence of `c` on the **right side** and then store the smallest out of the two in an `output` array.


## Detailed Explanation:

Pretty simple approach.

We loop through `s` in two passes (i.e. two times).

In the first pass we calculate the distance of the current character in `s` to the closest occurence of `c` on the **left hand side**.


In the second pass we calculate the distance of the current character in `s` to the closest occurence of `c` on the **right hand side** AND then compare the left and right distances and store the smallest in an `output` array.





## Key Challenges:





```python
class Solution:
    def shortestToChar(self, s: str, c: str) -> List[int]:

        # initialize 'previous' to -infinity to represent no occurrence of 'c' to the left yet
        previous = float('-inf')

        # initialize the output array to store the shortest distances
        output = []

        # loop through every character in s and their corresponding indices
        # and calculate their distance from the closest 'c' on the left side
        for index, char in enumerate(s):
        
        # check if character equals c
            if char == c:
                # if the current character is 'c', update 'previous' to the current index
                previous = index
        
        # calculate the distance from the closest 'c' on the left and add to output
            output.append(index - previous)
        
        # initialize 'previous' to positive infinity to represent no occurrence of 'c' to the right yet
        previous = float('inf')

    # loop through every character in s from end to start / right to left
    # and calculate the distance from the closest 'c' on the right side
        for index in range(len(s) - 1, -1, -1):

            # check if current character equals c
            if s[index] == c:
                
                # update previous with index of current `c` character in s
                previous = index
            
    # update the output with the minimum distance from either left or right 'c'            
            output[index] = min(output[index], previous - index)
        
        # return the final output array containing the shortest distances to 'c'
        return output
    
    
    
## Time complexity: O(n)
## Space Complexity: O(n)
```
