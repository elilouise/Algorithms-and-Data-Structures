## Approach Overview:

Loop through the string from both ends to form the longest **equal (non-empty) prefix** and the longest **equal (non-empty) suffix** we possibly can. 

Then check if they're both made up of the same character (i.e. if they're equal), if yes then delete them from `s`. 

If not then stop this process altogether and return the minimum length of string remaining after the deletion of similar ends (similar prefix and suffix).

## Detailed Explanation:

I really like these string manipulation questions using two pointers. And this one was no exception.

So the idea is the form as long of an **equal prefix and suffix** possible by scanning through the characters of `s` from both ends (in both the forward and reverse direction). 

After we finish building the prefix and suffix, we then check if they're both made up of the same character.

- If yes we delete them from `s` and continue the process (looking for equal prefix and suffixes to delete). 

- If not, we terminate and then reurn the length of the remaining strings (we stop once we find a mismatch).

**N.B: We need to ensure that the prefix and suffixes never intersect at any index the process of building prefix and suffixes.**

## Solution:

```python
class Solution:
    def minimumLength(self, s: str) -> int:

        left, right = 0, len(s) - 1

        minLength = right - left + 1
        
        # enter loop if leftmost character equals rightmost character 
        # AND if left pointer has NOT  met / intersected / crossed right pointer  
        while s[left] == s[right] and left < right:
            
        # enter loop if the next upcoming left character equals the current left character 
        # AND if the left pointer doesn't intersect with the current right pointer after we increment it
        # basically we're trying to find the longest EQUAL prefix possible
            while s[left] == s[left + 1] and left + 1 < right:
                # increment left to look at the next left character 
                left += 1
            
            # enter loop if next right character equals the current right character
            # AND if the right pointer doesn't intersect the left pointer after we decrement it
            while s[right] == s[right - 1] and left < right - 1:
                # decrement right pointer to look at next right character
                right -= 1

            # increment left (to "delete" current prefix)   
            left += 1

            # increment right (to "delete" current suffix)
            right -= 1

            # calculate the length of resulting string after we've deleted similar ends
            # and update `minLegnth` if its the smallest one we've discovered so far
            minLength = min(minLength, right - left + 1)
            
        # return the minimum length of string after we've deleting similar ends
        return minLength
```

## Alternative version of the deleting matches and also checking for mismatch:

```python
class Solution:
    def minimumLength(self, s: str) -> int:

        left, right = 0, len(s) - 1

        # loop until left and right pointer meet  
        while left < right:

            # check if current left character doesn't equal current right character
            if s[left] != s[right]:
                # terminate loop 
                break
          
            # loop until the next left index intersects the position of the right pointer
            # and until the current left character equals the next left character 
            while left + 1 < right and s[left] == s[left + 1]:
                # increment left
                left += 1
            
            # loop until the next right index intersects the position of the left pointer
            # and until the current right character equals the next right character
            while right - 1 > left and s[right] == s[right - 1]:
                #decrement right
                right -= 1
            
            # increment left
            left += 1
            # decrement right
            right -= 1
        
        # return length of remaining strings 
        return right - left + 1
```
