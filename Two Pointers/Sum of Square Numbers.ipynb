# 633. Sum of Square Numbers


### Difficulty: <font color = orange> Medium </font>

---

Given a non-negative integer `c`, decide whether there're two integers `a` and `b` such that $a^2 + b^2 = c$.

## Approach Overview:

Iterate through the range of `a` and `b` values and check if $a^2 + b^2 = c$. If it adds up then return <font color = green>`True`</font>. 

Else if $a^2 + b^2 > c$ then decrement `b` (i.e. right pointer).

## Detailed Explanation:

We need to realize that `a` and `b` will fall in the range of 0 to **int($\sqrt(c)$)**.

<i> $b = int(\sqrt(c))$ </i>

<b> code implementation of `a` / `b` range </b>

`b = right = int(sqrt(c))` # algorithmical implementation

Once we have the range sorted then we simply iterate through it and check whether it adds to the desired result. 

<b> code implementation of iterating through `a` / `b` range </b>


`for left in range(right + 1):` # iterate through 0 to b

If it's greater than the target value (i.e. `c`), then we decrement `b` (i.e. `right` pointer). 


## Key Challenges:

Figuring out the range of `a` and `b`.

## Solution:

```python
class Solution:
    def judgeSquareSum(self, c: int) -> bool:

        # initialize right pointer
        # e.g: if c = 5 then 
        # int(sqrt(c)) = 2
        right = int(sqrt(c)) 
        
        # loop from 0 to right (int(sqrt(c)))
        for left in range(right + 1):
            
            # check if left^2 + right^2 > c 
            if left**2 + right**2 > c:

                # if true (meaning sum is too large) then decrement `right`
                right -= 1
            
            # check if left^2 + right^2 = c
            if left**2 + right**2 == c:
                # if true, return True
                return True
        
        # no sum exists so return False
        return False            
```
