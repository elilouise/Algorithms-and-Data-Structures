## Approach Overview:

Remove all redundant spaces from `s` (leading, trailing & excessive ones between words). Then reverse the positions of all individual characters in `s`, and finally reverse them into words.

## Detailed Explanation:

This one was for the books. We had to perform a set of elaborate operations on this string to format `s` in the correct sequence to obtain the answer.

Firstly, we had to remove all redundant spaces from `s`. Redundant includes all leading, trailing and excessive spaces between words (trimming down multiple spaces between two words to a single space).

We use a helper function to do this. 

Secondly, we then reverse each character in the `s`, again we use a helper function to help us achieve this.

Lastly, we then loop through every character in `s` but this time we only reverse all the characters in every **word** we come across in `s`.

Finally, we convert the list into a string and return the final output.

## Solution:

```python
class Solution:

    def reverseEachWord(self, string):
        'helper function to reverse each word in string'

        # initialize start to 0 (i.e. at first word character in string)
        start = 0

        # loop through every character in string
        for end in range(len(string) + 1):

            # check if current character is a space or if we've exceed the bounds of string
            if end == len(string) or string[end] == " ":
                # reverse the word 
                self.reverseString(string, start, end - 1)
                # update `start` to point at the starting character of the next word in `string`
                start = end + 1    
            
    def trimSpaces(self, string):
        'helper function to remove redundant / extra spaces in string'

        # initialize left to point at first character in `string`
        # intialize right to point at last character in `string`
        left = 0
        right = len(string) - 1
        
        # remove leading spaces in string
        # loop if currrent leftmost character is empty space
        while string[left] == " ":

            # increment left to point at next left character in string
            left += 1

        # remove trailing spaces in string
        # loop if current rightmost character is empty space
        while string[right] == " ":

            # decrement right to point at the next rightmost character in string
            right -= 1

        # initialize array that will store the formatted string 
        # formatted string contains only a single space between words and has no leading & trailing spaces 
        outputString = []

        # loop through every character in string
        for index in range(left, right + 1):

            # check if current character is an english letter (i.e. it's NOT a space)
            if string[index] != " ":
                # add the letter to `outputString`
                outputString.append(string[index])

            # check if the last character in `outSputString` is not space    
            #elif outputString[-1] != " ": 
                # add the space to `outputString` (so we can have a single space between words) 
                #outputString.append(string[index])
            else: 
                # check if the last character added in `outputString` was a space 
                if outputString[-1] == " ": 
                # skip the current character (space) to prevent adding multiple spaces between words
                    continue

                # last character added was NOT a space (i.e. it was a letter)
                else:
                    # add the space to `outputString` (so we can have a single space between words) 
                    outputString.append(string[index])
        
        return outputString

    def reverseString(self, string, l , r):
        'helper function to reverse the characters in the input string'

        # continue until l and r pointers meet
        while l < r:

            # swap the current left and right characters 
            string[l], string[r] = string[r], string[l]

            # increment l to look at next left character
            l += 1 

            # decrement r to look at next right character 
            r -= 1
            
        
    def reverseWords(self, s: str) -> str:

        s = list(s)
      
        # reverse all characters in string s
        self.reverseString(s, 0, len(s) - 1)  

        # trim redundant / extra space in string
        s = self.trimSpaces(s)
        
        # reverse each word in string s
        self.reverseEachWord(s)
        
        # convert the output (reversed words) from list to string
        reversed = "".join(s)

        return reversed
```
