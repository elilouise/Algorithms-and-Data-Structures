## Approach Overview

We iterate through every elements in `nums` using two pointers to 'remove' all occurences of `val`.

We achieve this using two pointers (one for reading elements and one for writing elements).

Each time we encounter an element that's not `val` we write its value to the left pointer's (read pointer) position, else we skip over the element and read the next available one (since current one is equal to `val`).

This results in the first `k` elements of nums containing elements which are not equal to `val`.


## Detailed Explanation:

We employ a two pointer technique to remove all occurences of `val` in `nums` in place, such that the first `k` elements of `nums` contains elements which are not equal to `val`.

We set one pointer as the **read pointer**. It's job is to solely read elements.

And we set the other as the **write pointer**. It's job is to selectively write elements.

How does it work?

Well initially both pointers are set to the beginning of `nums`.

We read the current element using the read pointer and check its value.

If the value equals `val`, we only increment the read pointer to look at the next element and leave the left writer pointer position unchanged.

If the value does not equal `val`, we write the current read pointer element into the write pointer position and then increment both the read and writer pointer.


Since the `left` / writer pointer only moves / increments if we find a desired element (i.e. elements not equal to val) and move it to the front. The position of left pointer after we've traversed through `nums` should be equal to `k` (i.e. it should be pointing to the `kth` element that's not equal to `val` in `nums`).

## Solution:

```python
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int: 
        # initialize left to point at first element in `nums`
        left = 0
        
        # loop through all element in `nums`
        for right in range(len(nums)):
           
            # check if current element is not equal to val
            if nums[right] != val: 
                # place the current element at the left pointer 
                nums[left] = nums[right]
                # move the left pointer to the next position
                left += 1
        
        # return k (number of elements in `nums` that's not equal to val)
        return left
```
