# 917. Reverse Only Letters

### Difficulty: <font color = green> Easy </font>

## Approach Overview:

Loop through the string using two pointers in search for pairs of english letters to reverse. Each time we encounter a non-english letter, we continuously skip it and move to the next character to find an english letter. 



## Detailed Explanation:

The idea is simple, we go through characters in string `s` using two pointers. 

At each iteration we compare a pair of characters and check if they're english letters. If they are, we reverse their positions, if not we ignore them by skipping them and moving to the next available characters until we find english letters. 

We use a helper function to help us figure out if a character is an english letter.

Inorder to manipulate the characters in the string, we need to convert it to a list because strings are immutable in Python (i.e. they cannot be changed).


## Key Challenges:



## Code Solution:

```python
class Solution:
    def isEnglishLetter(self, char):
        'helper function to help us figure out if the input character is an english letter'
        'INPUT: char (input character)'
        'OUTPUT: Boolean (True if character is an english letter, otherwise False)'
        
        # check if char (input character) is either a small or big english letter 
        # return True if it is. Otherwise False.
        return 65 <= ord(char) <= 90 or 97 <= ord(char) <= 122


    def reverseOnlyLetters(self, s: str) -> str:
        # convert string to list 
        s = list(s)
        # initialize left and right pointer
        l, r = 0, len(s) - 1
        
        # continue until pointers meet
        while l < r:
            # check if left character is not an english letter
            while l < r and self.isEnglishLetter(s[l]) == False:
                # increment left until current left character is an english letter
                l += 1
            # check if right character is not an english letter
            while l < r and self.isEnglishLetter(s[r]) == False:
                # decrement right until current right character is an english letter
                r -= 1          
            # check if both left and right character are english letters
            if self.isEnglishLetter(s[l]) and self.isEnglishLetter(s[r]):
                # swap left and right characters
                s[l], s[r] = s[r], s[l]
                # increment left pointer to look at next left character
                l += 1
                # decrement right pointer to look at next right character
                r-= 1
                
        # convert list into a string and output it        
        return "".join(s)        
              
```
