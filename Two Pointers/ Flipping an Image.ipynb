# 832. Flipping an Image

### Difficulty: <font color = green> Easy </font>

---


## Approach Overview:

Iterate through each element in `image` sequentially. First scan through each row and flip the image horizontally with the help of two pointers. Then loop through every element (bit) and invert it.

## Detailed Explanation:

Pretty straightforward solution, we only have to do things in the correct sequence (as is often the case when it comes to dealing with arrays). 

First off we flip the image horizontally row by row using two pointers. 

We initialize them to point at the leftmost and rightmost bits / elements in a row and then swap / flip their positions. After we then move them towards each other until they meet.

Finally all that is left is inverting every bit / element. We can easily do that we the help of `for loops`. So we loop through every bit / element in `image` and invert them. 


## Key Challenges:

1. Surprisingly I managed to do this on my own. Guess the only slight challenge for to prevent the pointers from exceeding the bounds of `image` matrix.

## Solution:

```python
class Solution:
    def flipAndInvertImage(self, image: List[List[int]]) -> List[List[int]]:
        
        n = len(image)
        
        # loop through every rows
        for row in range(n):
            
            # initialize left (first bit in current row / leftmost bit)
            # initialize right (last bit in current row / rightmost bit)
            left, right = 0, n - 1

            # continue until left and right meet
            while left < right:
                
                # flip the left and right bit in current row
                image[row][left], image[row][right] == image[row][right], image[row][left]        
                # increment left pointer to process the next left bit in current row
                left += 1
                # decrement right pointer to process the next right bit in current row
                right -= 1
                
        # loop through every row in image     
        for row in range(n):
            # loop through every column in image
            for column in range(n):

                # check if current row element is equal to "1"
                if image[row][column] == 0:
                    # invert it to "0"
                    image[row][column] = 1
                else:
                    # invert it to "1"
                    image[row][column] = 0
        
        return image
```
