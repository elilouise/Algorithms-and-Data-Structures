# 5. Longest Palindromic Substring

## Approach Overview:

Iterate through each character in the string `s`, and for each character, check if it can be the center of an odd-length palindrome. If it can, expand outward from the center to find the longest palindrome. 

Then, ALSO check if there is a even-length palindrome centered between the current character and the next one, and expand around this center to find the longest palindrome. Track and update the longest palindrome found during this process.

## Detailed Explanation:

To solve this problem, we need to scan through each character in `s` and **recognize** that there are **two types of palindromes we need to check for.** 

<u>1st case:</u>

An **odd-length palindrome**, where the left and right characters around the center always mirror each other. 



Once we identify an **odd-length palindrome** within `s`, we continuously expand around the center of the current character to find the longest palindrome possible.

<u>2nd case:</u>

An **even-length palindrome**, where the current character (left) and the next one (right) are the same. 



We treat the space between these two characters as the center and expand outward, checking if the characters on the left and right continue to mirror each other. This process continues until no more matching characters are found, giving us the longest even-length palindrome.

We continually scan through each character in `s` check for both **odd-length and even-length** palindromic substrings at each character until we reach the end of string `s`.

## Key Challenges:

1. **I struggled to implement a robust method for checking both odd-length and even-length palindromic substrings. While I understood that I needed to compare the left and right characters for odd-length cases, and the left and right + 1 characters for even-length cases, I couldn't translate that logic correctly into code.**

2. **I also had difficulty writing code to accurately calculate and store the longest palindrome found in the string s.**

## Solution:

```python
class Solution:

    def longestPalindrome(self, s: str) -> str:
        
        # initialize a variable to store the longest palindrome we find in s
        longestPalindrome = ""

        # initialize a variable to store the length of the longest palindrome
        lengthOfPalindrome = 0
        
        # loop through each character in s
        for index in range(len(s)):

            # We now check for odd-length palindrome at current character #

            # initialize left and right to index (of the current character in s)
            left = right = index
            

            # continue as long as both left and right characters are within bounds of s 
            # and matches each other 

            while left >= 0 and right < len(s) and s[left] == s[right]:

                # check if length of current palindrome > length of longest palindrome we've previously found
                if right - left + 1 > lengthOfPalindrome:
                    # store the current longest palindrome 
                    longestPalindrome = s[left: right + 1]
                    # store the length of current longest palindrome 
                    lengthOfPalindrome = right - left + 1


                # decrement left and increment right to search for an even longer odd-sized palindrome
                left -= 1
                right += 1

            
            # We now check for an even-length palindrome at current character #

            # initialize left to `index` and right to `index + 1` (for the current character)
            left, right = index, index + 1

            # continue as long as both left and right characters are within bounds of s 
            # and matches each other
            while left >= 0 and right < len(s) and s[left] == s[right]:
                
                # check if length of current palindrome > length of longest palindrome we've previously found
                if right - left + 1 > lengthOfPalindrome:
                    # store the current longest palindrome 
                    longestPalindrome = s[left: right + 1]
                    # store the length of current longest palindrome 
                    lengthOfPalindrome = right - left + 1

                # decrement left and increment right to search for an even longer even-sized palindrome
                left -= 1
                right += 1

        # return the longest palindromic substring in s
        return longestPalindrome
```
