# 345. Reverse Vowels of a String

### Difficulty: <font color = green> Easy </font>

---



## Approach Overview:

Loop through `s` from both ends in search for vowels, once we encounter them, we swap / reverse their positions

## Detailed Explanation:

Pretty straightforward, we scan through `s` using two pointers running through both ends of `s`. We do this until we find a pair of vowels (from the left and the right), once we do we then swap / reverse their positions. 

Important to note that we need to convert `s` from string to array (list) to be able to achieve this because strings are immutable in python!

Once we finish over vowel swapping operation we then convert the `s` array (list) to a string again and then return the final result.

```python
class Solution:
    def reverseVowels(self, s: str) -> str:

        # initialize left pointer
        left = 0
        # initialize right pointer
        right = len(s) - 1
        
        # convert s from string to list
        s = list(s)
        
        # initialize to set() to store all vowels
        vowels = set('AEIOUaeiou')
        
        # loop until left and right pointer meet
        while left < right:

        # enter loop if current left character is not a vowel 
        # and if left pointer hasn't yet intersected right pointer
            while s[left] not in vowels and left < right:
                # increment left pointer
                left += 1

        # enter loop if current right character is not a vowel 
        # and if left pointer hasn't yet intersected right pointer
            while s[right] not in vowels and left < right:
                # decrement right pointer
                right -= 1

        # swap current left and right characters (vowels)
            s[left], s[right] = s[right], s[left]
        # increment left pointer   
            left += 1
        # decrement right pointer
            right -= 1
            
        # convert list to string 
        s = "".join(s) 
        
        # return s
        return s 
```
