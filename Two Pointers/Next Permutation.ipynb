# 31. Next Permutation

## Approach Overview:

To find the next permutation of `nums` we iterate through its integer elements from end to start until we find the first **non-increasing / decreasing** integer element. 

We then iterate through `nums` from end to start until but this time to find the integer element that's just larger the previously identified **non-increasing / decreasing** integer element. Once we do, we then swap their positions and reverse the remaining elements on the right.


## Detailed Explanation:

<font color = navy> Identify the First Non-Increasing Element from the End: </font> 

Start from the end of the list and move backwards to find the first element that is not in increasing order. 

**This element is identified as the first element that is smaller than its next element.**

<font color = navy>Find the Element Just Larger than the Identified Element: </font> 

Again, start from the end of the list and move backwards to find the first element that is larger than the identified element from step 1.

<font color = navy>Swap the Two Elements:</font> 

Swap the identified element from step 1 with the element found in step 2.

<font color = navy> Reverse the Sublist: </font> 

Reverse the sublist that starts just after the element identified in step 1 to the end of the list. This ensures that we get the smallest lexicographical order for the permutation.

```python
class Solution:

    def reverse(self, arr, l, r):

        while l < r:
            arr[l], arr[r] = arr[r], arr[l]
            l += 1
            r -= 1
      
    def nextPermutation(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        left = len(nums) - 1
        # iterate through the digits in nums (from end to start) until we the first non-increasing / pivot digit
        while left - 1 >= 0 and nums[left - 1] >= nums[left]:
            left -= 1

        # check if left = 0 (all the digits in nums are arranged in descending order)
        if left == 0:
            # in this case no next larger permutation is possible
            # so nums must be rearranged  as the lowest possible order (i.e., sorted in ascending order)
            self.reverse(nums, left, len(nums) - 1)
        
        else:

            # initialize index to the end of nums array
            index = len(nums) - 1 

            # iterate through nums until we find the first digit that's bigger than the non-increasing / pivot digit 
            while nums[left - 1] >= nums[index]:
                # decrement index to look at next rightmost digit in nums
                index -= 1

            # swap pivot digit with the (first) bigger digit on the right
            nums[left - 1], nums[index] = nums[index], nums[left - 1]
            
            # reverse the segment of nums to the right pivot digit
            self.reverse(nums, left, len(nums) - 1)
```
