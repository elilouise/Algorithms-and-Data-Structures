## Approach Overview:

Using two pointers, traverse through the string in search of a pair of mismatched characters. If we find one then check if after 'deleting' either the left or right character from it the resulting string becomes a palindrome. 

## Detailed Explanation:

The core idea is to traverse through the string and process a pair of characters at each turn. We use two pointers to facilitate this, one that tracks the leftmost character and one tracking the rightmost character. 

If we encounter a pair of characters that doesn't match, we check if after deleting either the left OR the right character, the string becomes a palindrome. 

- If it becomes a palindrome after the deletion, we return **True**

- If not, we return **False**

We use a helper function to help us check whether a string is a palindrome.

## Key Challenges:

1. Finding out how to 'delete' a character from the string without actually deleting it. 

P.S: We delete a character by simply skiping it. We do this by either incrementing or decrementing the left or right pointer. That's how we 'delete' a character

## Solution:

```python
class Solution:

    def isPalindrome(self, l, r, string):
        ' helper function to check if input string is a palindrome'
        
        # continue until left and right pointers meet / intersect
        while l < r:

            # check if current left character is not the same as current right character
            if string[l] != string[r]:
                # return False (since string cannot be a palindrome)
                return False
            # increment left pointer to look at next left character in string    
            l += 1
            # decrement right pointer to look at next right character in string
            r -= 1    

        # return True (since we know `string` is a palindrome)    
        return True        

    def validPalindrome(self, s: str) -> bool:
        # initialize left and right pointers
        left, right = 0, len(s) - 1

        # loop until left and right pointers meet / intersect 
        while left < right:

            # check if left character is not the same as right character
            if s[left] != s[right]:
                
                # check if after the deleting either the current left character or right character, the resulting string becomes a palidrome   
                if self.isPalindrome(left + 1, right, s) or self.isPalindrome(left, right - 1, s):      
                    # return True
                    return True
                #otherwise 
                else:
                    # return False
                    return False    
            
            left += 1
            right -= 1

        return True    
```
