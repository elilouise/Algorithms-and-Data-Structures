# 2000. Reverse Prefix of Word

### Difficulty: <font color = orange> Medium </font>

---



## Approach Overview:

Pretty straightforward. Find the index position of the first occurrence of `ch` in word and reverse the segment of `word` that starts at index `0` and ends at the index of the first occurrence of `ch` (inclusive) and return the result. If `ch` does not exist in word then return the original word.

## Detailed Explanation:

Again pretty straightforward.

Firstly we need to locate the index of the first occurrence of `ch` in word.

Then we use a helper function (to make things as clean as possible) to reverse the segment of `word` that starts at index `0` and ends at the index of the first occurrence of `ch`.

We then return the result. If `ch` does not exist in word then return the algo will just return the original word.

## Solution:

```python
class Solution:

    def reverse(self, string, i):
        'helper function to reverse the segment / susbtring of `word` that starts at 0 and ends at index (inclusive)'

        # initialize l to beginning position of `string`
        l = 0
        # initialize r to end position of `string`
        r = i

        # continue until l and r pointers meet
        while l < r:
            # swap the left and right characters
            string[l], string[r] = string[r], string[l]
            # move right pointer to look at next rightmost character in `string`
            r -= 1
            # move left pointer to look at next leftmost characater in `string`
            l += 1

    def reversePrefix(self, word: str, ch: str) -> str:

        # find the index of the first occurence of `ch` in `word`
        index = word.find(ch)

        # check if `ch` is not in `word`
        if index == - 1:

            # return original string (i.e. word)
            return word

        # convert word from string to list
        word = list(word)   

        # reverse the segment / susbtring of `word` that starts at 0 and ends at index (inclusive)
        self.reverse(word, index)
        
        # convert word from list to string 
        return "".join(word)
```
