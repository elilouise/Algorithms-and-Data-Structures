## Approach Overview:

Using two pointers (one pointing to the beginning element and one to the end element in `nums`) loop through `nums`. 

At each iteration pick the largest out of the two elements and insert them into the largest available element index position in `output`. Continue until all index positions in `output` is filled.  


#### Refined version:

Using a two-pointer approach, traverse the array from both ends, comparing the absolute values of the elements and filling the `output` array from the largest to the smallest square. At each step, we compare the absolute values at these pointers and insert the square of the larger value into the next position in the output array, starting from the end. 

This ensures that the final array is sorted.


## Detailed Explanation:

This approach leverages the sorted nature of the input array, which may contain both negative and positive numbers.

We begin by initializing an output array of the same length as the input array, filled with zeros. We also set two pointers: one at the start (left) and one at the end (right) of the input array. Additionally, we maintain an index pointer at the end of the output array, which helps us place the squares in descending order.

The core of the algorithm involves iterating through the input array using the two pointers. 

In each iteration, we compare the absolute values of the elements pointed to by the left and right pointers. Since squaring a larger absolute value results in a larger number, we place the larger square at the current index position in the output array.

If the absolute value at the left pointer is greater than that at the right, we square the left value and insert it into the output array at the current index. 

We then move the left pointer one step to the right. If the absolute value at the right pointer is greater or equal, we square the right value and place it in the output array, then move the right pointer one step to the left. 

After placing the square in the output array, we decrement the index pointer to fill the next position.

This process continues until the left pointer surpasses the right pointer. At this point, all elements in the input array have been processed, and the output array is fully populated with squared values in non-decreasing order.

Finally, we return the output array as the result.

## Solution:

```python
class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        # output array to store square of elements in a sorted (ascending) order
        output = [0] * len(nums)

        # initialize index to point to end position in `output`
        index = len(nums) - 1
        
        # initialize left pointer to start position in `nums`
        left = 0

        # initialize right pointer to end position in `nums`
        right = len(nums) - 1

        # continue until left and right pointers fully cross each other 
        while left <= right:

            # check if left element is greater than right element 
            if abs(nums[left]) > abs(nums[right]):
                # insert left element in current index position in `output`
                output[index] = (nums[left]) ** 2
                # increment left pointer to look at next left element in `nums`
                left += 1
            else:
                # insert right element in current index position in `output`
                output[index] = (nums[right])**2
                # decrement right pointer to look at next right element in `nums`
                right -= 1
            # decrement index to populate `output` with next largest element in `nums`
            index -= 1        
        
        # return output (array of the squares of each number sorted in non-decreasing order)
        return output
```
