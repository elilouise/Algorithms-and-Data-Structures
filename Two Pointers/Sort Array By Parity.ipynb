## Detailed Explanation:

We'll maintain two pointers `left` and `right` to facilitate the sorting. 

The idea is everything below `left` / left side has parity **0** (ie. `nums[k] % 2 == 0` when `k < right`), and everything above `right` has parity **1**.

Then, there are 4 cases for (`nums[left] % 2`, `nums[right] % 2`):

- If it is (0, 1), then everything is correct: `left++` and `right--`.


- If it is (1, 0), we swap them so they are correct, then continue.


- If it is (0, 0), only the `left` place is correct, so we `left++` and continue.


- If it is (1, 1), only the `right` place is correct, so we `right--` and continue.


We will loop through `nums` to reposition / swap all integers based on their respective parity until its all sorted.

```python
class Solution:
    def sortArrayByParity(self, nums: List[int]) -> List[int]:
        # initialize left pointer to beginning of `nums`
        left = 0
        # initialize right pointer to end of `nums`
        right = len(nums) - 1

        # continue to loop until left and right pointer meet
        while left < right:

            # check if leftmost integer is odd and if rightmost integer is even
            if nums[left] % 2 != 0 and nums[right] % 2 == 0:
                # swap the positions of the left and rightmost integers
                nums[left], nums[right] = nums[right], nums[left]
                # increment left to look at next leftmost integer in `nums`
                left += 1
                # decrement right to look at next rightmost integer in `nums`
                right -= 1
            
            # check if leftmost integer is odd and if rightmost integer is odd
            elif nums[left] % 2 != 0 and nums[right] % 2 != 0:
                # decrement right to look at next rightmost integer in `nums`
                right -= 1
            
            # check if leftmost integer is even and if rightmost integer is odd
            elif nums[left] % 2 == 0 and nums[right] % 2!= 0:
                # increment left to look at next leftmost integer in `nums`
                left += 1
                # decrement right to look at next rightmost integer in `nums`
                right -= 1
            
            # leftmost integer is even and rightmost integer is even
            else:
                # increment left to look at next leftmost integer in `nums`
                left += 1

        # return array sorted by parity
        return nums
```
