## Approach Overview:

Loop through the elements of `arr` in descending order (starting with the largest one) and flip it to the correct position (this can be done in either one or two flips **at most**). Each time we perform a flip we make sure to record it to an output array. 

## Detailed Explanation:

Very weird question to be honest but anyway here's the break down.

We loop through the elements of `arr` in descending order (starting with the largest element).

For each element, we flip it to its correct position. This can be done in either one or two flips at most:

1. If the element is already in the correct position, no flips are needed.

2. If the element is not in the correct position:

- Flip it to the front of the array (if it's not already there), and record the flip.

- Then flip it from the front to its correct position, and record the flip.

Record the "flip index" (i.e., the position up to which the flip is made, denoted as `k`) each time a flip is performed, and add this to an output array.

Repeat until all elements except the smallest one (1) are in their correct positions. By this point, the array will be fully sorted, so there's no need to flip the smallest element.

## Solution:

```python
class Solution:
    def reverse(self, array, k, results):
            
            l = 0
            r = k - 1

            # reversed all elements in arr located from 0th to k-1th position
            while l < r:

                array[l], array[r] = array[r], array[l] 
                r -= 1
                l += 1   
            
            # save the k value in `output` 
            results.append(k)

    def pancakeSort(self, arr: List[int]) -> List[int]:
        output = [] # initialize array to store k values 

        valueToSort = len(arr) # pick the largest element in the `arr`
        
        for value in range(valueToSort, 0, -1):
            
            # check if `valueToSort` is NOT in correct position (i.e. it's not in rightmost position)
            # locate current position of `valueToSort` in `arr`
            index = arr.index(value)  
            
            if index != value - 1: 
                if index != 0:
                # if true, then flip it to the correct position 
                    self.reverse(arr, index + 1, output)
                
                self.reverse(arr, value, output)

        return output
```

```python
class Solution:
    def reverse(self, input_arr, output, num):

        # define the correct index position for the current value
        # (i.e. define the index at which the current value should be placed at)
        right = num - 1
        # locate current index of current value (i) in `arr`
        index = input_arr.index(num)

        # check if the value is already in starting position in `arr`)
        if index == 0:  

            # flip it to the correct position
            self.flip(input_arr, 0, right)
            # add the value (k = value) to `output` 
            output.append(num)
            
        else:

            # add k (= index + 1) to `output`
            output.append(index + 1)
            # flip the value to the beginning position in `arr`
            self.flip(input_arr, 0, index)
            
            # add k (= num) to `output`
            output.append(num)
            # flip the value to its correct position 
            self.flip(input_arr, 0, num - 1)
            
    def flip(self, ar, l, r):
        'helper function to reverse current segment, arr[l:r] of array'

        # loop until l and r pointers meet 
        while l < r:
            # swap the positions of the left and right elements in `arr`
            ar[l], ar[r] = ar[r], ar[l]
            # increment l pointer
            l += 1
            # decrement r pointer
            r -= 1
    
    def pancakeSort(self, arr: List[int]) -> List[int]:
       
        n = len(arr)
        output = []

        for currentValue in range(n, 1, -1 ):

            if arr.index(currentValue) == currentValue - 1:
                continue

            self.reverse(arr, output, currentValue)

        return output
```
