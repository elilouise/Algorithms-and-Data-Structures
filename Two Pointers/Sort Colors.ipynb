## Approach Overview:

Count the number of occurrences of each element in `nums` and reposition each element in ascending order by using the occurrence count.  

## Detailed Explanation:

This is a clever technique. The idea is to count the occurrences of all elements in `nums` and store it in a dictionary. 

We then loop through each element in ascending order and re-position them in the correct order in `nums`. Since we know how many times they occur we can thus know how to correctly position them in `nums`.

Take `nums = [2, 0, 2, 1, 1, 0]` as an example.

After we count the occurences of each element, the dictionary will have the following results. 

`countOfColors = {2: 2, 0: 2, 1: 2}`

2 appears 2 times, 0 appears 2 times and 1 appears 2 times.

To sort `nums` in ascending order we'll first need to put all **0** in the first positions, then **1** and lastly **2**.  

And since we know how many **0**s, **1**s and **2**s there are we can easily position each of them in `nums` ascending.

For example here we know that **0** needs to be inserted in the first two beginning positions and then **1** needs to be inserted in the next two preceeding positions and finally **2** needs to be inserted in the last two remaining positions in `nums`.


We use this clever implementation to achieve this.

## Alternative Solution:

## Approach Overview:

Loop through `nums` using three pointers (one to track left & right element and one to track the current element in `nums`). 

Each time current element equals `0` we swap it with the left pointer element and when current equals `2` we swap it with right pointer element. 

## Detailed Explanation:

This was an interesting technique to implement, we essentially relocated all `0`s to the begining of `nums` and all `2`s to the end of `nums`. Notice this means all the `1`s will settle in the middle.

I think this only works for the special case of three distinct numbers.

Anyway, let's go into the details.

We employ three pointers to implement the sorting. 

The left pointer will track the position of the left elements (where we will insert `0s`) and the right pointer will the track the position of the right elements (where we will insert `2s`), and finally index will track the current element in `nums`.

If current element = `0`, we swap it with the element located at the `left` pointer and then increment both `index` and `left` pointer.

If current element = `2`, we swap it with the element located at the `right` pointer and then decrement the `right` pointer.

If current element = `1`, we just increment `index`.

## Solution:

```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """

        # initialize index (to keep track of current element in nums) 
        index  = 0
    
        # initialize left and right to begining and end element in nums
        left, right = 0, len(nums) - 1
    
        # loop until index crosses the right pointer
        while index <= right:
      
            # check if current element is equal to 0
            if nums[index] == 0:
        
            # swap left element with current index element   
                nums[left], nums[index] = nums[index], nums[left]
            # increment left pointer
                left += 1
            # increment index pointer 
                index += 1
      
            # check if current element is equal to 2 
            elif nums[index] == 2:
        
            # swap right element with current index element
                nums[right], nums[index] = nums[index], nums[right]
            # decrement right pointer
                right -= 1
            
            # else (nums[index] = 1)
            else:
                # increment index pointer
                index += 1  
```
