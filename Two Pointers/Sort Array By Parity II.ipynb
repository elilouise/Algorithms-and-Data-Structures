# 922. Sort Array By Parity II

### Difficulty: <font color = green> Easy </font>

----



## Approach Overview:

Using two pointers, one to track **even indexes** and one for **odd indexes**, loop through `nums` array until we discover misplaced elements (e.g: an odd element at an even index or an even element at an odd index) and correct any misplaced elements by swapping them into their correct positions.

## Detailed Explanation:

We initialize two pointers, `even` and `odd`. 

The even pointer starts at index `0` and is responsible for tracking even indices. The odd pointer starts at index `1` and is responsible for tracking odd indices.

Loop through `nums` array until we discover misplaced elements. 

If the element at the `even` index is correctly placed (i.e., an even number), we move the even pointer to the next even index by incrementing it by 2.

Similarly, if the element at the odd index is correctly placed (i.e., an odd number), we move the odd pointer to the next odd index by incrementing it by 2.

When we find an even index containing an odd number and an odd index containing an even number, we swap these elements to correct their positions.

The process continues until all misplaced elements are swapped to their correct positions.

## Key Challenges:

1. Ensuring that the pointers remain within the bounds of the array while checking and swapping elements.


2. Efficiently identifying misplaced elements and swapping them using two pointers (one for even and one for odd indices)

## Solution:

```python
class Solution:
    def sortArrayByParityII(self, nums: List[int]) -> List[int]:
        
        # initialize pointer for even index
        even = 0
        # initialize pointer for odd index
        odd = 1
        
        # continue as long as even and odd pointer are within the bounds of `nums`
        while even < len(nums) and odd < len(nums):
            
            # loop if even pointer is within `nums` bounds and if current value is an even number
            while even < len(nums) and nums[even] % 2 == 0:
                # move to the next even index in `nums`
                even += 2
            
            # loop if even pointer is within `nums` bounds and if current value is an even number
            while odd < len(nums) and nums[odd] % 2 != 0:
                # move to the next odd index in `nums`
                odd += 2
            
            # check if even and odd pointers are within the bounds of `nums`
            if even < len(nums) and odd < len(nums): 
                # swap even number at odd index with the odd number at the even index
                nums[even], nums[odd] = nums[odd], nums[even]
            
            # move to the next even index in `nums`
            even += 2
            # move to the next odd index in `nums`
            odd += 2
          
        return nums        
```
