## 217. Contains Duplicate

### Difficulty: Easy

Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

 

Example 1:

Input: nums = [1,2,3,1]
Output: true


Example 2:

Input: nums = [1,2,3,4]
Output: false


Example 3:

Input: nums = [1,1,1,3,3,4,3,2,4,2]
Output: true
 

Constraints:

$  1    <=  nums.length <= 105 $

$ -109  <=  nums[i]     <= 109 $

### Approach 1: Brute Force 

#### Check all possible pair of element combinations

Iterate through 'nums' array and compare each element against every other element in 'nums' array. 

If a value is detected twice then stop loop and return <font color = blue > True </font>.

Else return <font color = red > False </font>.


Time Complexity: $O(n^2)$

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        
        lengthOfNums = len(nums)
        
        for i in range(lengthOfNums-1):
            for j in range(i + 1, lengthOfNums):
                
                if nums[i] == nums[j]:
                    
                    return True
        
        
        return False
```

### Approach 2: Sorting

#### Sort array and check adjacent array elements for duplicates

Sort the 'nums' array in ascending order (smallest to biggest), scan through array once checking successive / adjacent elements for duplicates. 

Time Complexity: $O(nlogn)$ / Sorting has time complexity of $O(nlogn)$ (worst case scenario), n = length of 'nums' array

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:

        sortednums = sorted(nums)


        for i in range(len(nums)-1):

            if sortednums[i] == sortednums[i+1]:

                return True

        return False   
```

### Approach 3: Hash Set

#### Loop through array, store visited elements in a hash set and check if element already in hashset

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        
        seen = ()
        
        for num in nums:
            
            if num in seen:
                
                return True
            
            seen.add(num)
        
        
        return False
        
```
