# 2. Add Two Numbers

### Approach Overview:

Traverse both linked lists to extract their digits into strings (number1 and number2), reverse these strings to restore the original number order, and compute their sum. After obtaining the result, reverse the digits of the sum and convert them back into a linked list by creating new nodes for each digit.

### Detailed Explanation:

##### **Traverse the Linked Lists:**

Iterate through both linked lists (`l1` and `l2`), extracting each node's digit and appending it to two separate strings (`number1` for `l1` and `number2` for `l2`). This gathers the digits in reverse order since we are appending the least significant digits first.

##### **Reverse the Strings:**

After collecting the digits, reverse both `number1` and `number2` to obtain the correct number representation (most significant digit first).

##### **Add the Numbers:**

Convert the reversed strings into integers, then add them together to get the sum.

##### **Reverse the Sum:**

Convert the sum back into a string and reverse it to facilitate the construction of the linked list in the correct order, starting from the least significant digit.

##### **Rebuild the Linked List:**

Initialize a dummy node to simplify list construction. Loop through each digit of the reversed sum string, create a new node for each digit, and append it to the output linked list.

##### **Return the Result:**

Return the linked list starting from the first node after the dummy node, which represents the sum of the two input lists.

### Solution:

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        # variable to store all the digits in l1
        number1 = ""
        # variable to store all the digits in l2
        number2 = ""

        # loop through all the nodes in l1 and l2 
        while l1 or l2:

            # check if current node in l1 is not null
            if l1:
                # add the digit value of current node to "number1"
                number1 += str(l1.val)
                # move to next node in l1
                l1 = l1.next

            # check if current node in l2 is not null   
            if l2:
                # add the digit value of current node to "number2"
                number2 += str(l2.val)
                # move to next node in l2
                l2 = l2.next

        # reverse number1 and number2     
        number1, number2 = number1[::-1], number2[::-1]
        
        # Add the two numbers 
        output = int(number1) + int(number2)

        # convert the output sum to string
        output = str(output)

        # reverse the digits in output  
        output = output[::-1]

        # initialize a dummy node (to avoid NULL atrritube errors)
        outputNode = dummy = ListNode(-1)
        
        # loop through all the digits in output
        for digit in output:

            # create a new node and initialize it with the current digit
            outputNode.next = ListNode(int(digit))

            # move to the next node in output linkedlist
            outputNode = outputNode.next

        # return the summed linkedlist
        return dummy.next
```
