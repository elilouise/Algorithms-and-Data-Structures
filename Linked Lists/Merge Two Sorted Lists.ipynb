# 21. Merge Two Sorted Lists

## Approach Overview:

Traverse through both **list1** and **list2**. At each iteration, compare the current nodes from both lists, and add the node with the smaller value to the merged output list.

## Detailed Explanation:


The solution uses a single pointer to traverse and build the merged linked list while leveraging the pointers of the input lists to access and compare their current nodes. Here's a step-by-step breakdown:

#### Initialization:

Dummy Node: A dummy node is created to serve as the starting point of the merged list. This node helps in simplifying the edge cases, such as handling an empty list.

Pointer Initialization: Two pointers, current and output, are initialized to the dummy node. The current pointer is used to build the merged list, while the output pointer remains at the start of the list.

#### Traversal and Comparison:

We iterate through both list1 and list2 while neither list is empty. This ensures that we only process nodes that exist in both lists.

Comparison Logic: At each iteration, we compare the values of the current nodes in `list1` and `list2`.

If the value in `list1` is smaller or equal to the value in `list2`, link the current node of `list1` to the merged list and move the `list1` pointer to the next node.

Otherwise, link the current node of `list2` to the merged list and move the `list2` pointer to the next node.

Move Pointer: After linking a node to the merged list, move the current pointer to this newly added node.

#### Handling Remaining Nodes:

- `List1` Remaining Nodes: If `list1` still has nodes left after `list2` is exhausted, append all remaining nodes of `list1` to the merged list.

- `List2` Remaining Nodes: Similarly, if `list2` has nodes left after `list1` is exhausted, append all remaining nodes of `list2` to the merged list.

The head of the merged list is pointed to by output.next (skipping the dummy node). This is returned as the result.

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:

        # initialize output list with a dummy node
        output = current = ListNode(-1)

        # traverse both lists while neither is empty
        while list1 and list2:

            #check if the value of the current node in list1 >= list2
            if list1.val >= list2.val:

                # attach the current node from list2 to the merged list
                current.next = list2
                # move to next node in list2
                list2 = list2.next
            else:
                # attach the current node from list1 to the merged list
                current.next = list1
                # move to the next node in list1
                list1 = list1.next
           
            # move the current pointer to the newly added node
            current = current.next
       
        # check if we haven't reached the end of list1 
        if list1:
            # attach all remaining nodes in list1 to output list
            current.next = list1

        # check if we haven't reached the end of list2 
        elif list2:
            # attach all remaining nodes in list2 to output list
            current.next = list2

        # return the HEAD of the merged list (skipping the dummy node))
        return output.next
```
