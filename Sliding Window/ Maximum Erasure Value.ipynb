You are given an array of positive integers `nums` and want to erase a subarray containing unique elements. The score you get by erasing the subarray is equal to the sum of its elements.

Return *the **maximum score** you can get by erasing exactly one subarray*.

An array `b` is called to be a subarray of `a` if it forms a contiguous subsequence of `a`, that is, if it is equal to `a[l],a[l+1],...,a[r]` for some `(l,r)`.



## Approach Overview:

Iterate through every element in nums to search for the subarray in `nums` that has the maximum score. We use sliding window technique to implement this. 

At each iteration, we check if the current element can constitute a valid subarray (i.e. if we add it to our subarray, will it form a subgroup of **unique** & adjacent elements present in array `a`). If it's valid we add the element to the current subarray set, calculate the subarray's `score` & then expand the window. 

If not, we then start to remove elements from the current subarray window (starting with the leftmost element) until we successfully remove the duplicate element from the window (i.e. until window becomes valid again).

## Detailed Explanation:

In this problem, we are given an array of integers called `nums`. Our task is to find the subarray with the highest score, where the score is defined as the sum of the elements in the subarray. The subarray must also contain unique and contiguous elements.

**KEY NOTE: A subarray of `nums` is valid only if every element in it is unique and is contiguous / adjacent / neighbours each other in `nums` array.**

The approach:

Iterate through every element in nums to search for the subarray in `nums` that has the maximum score. We use sliding window technique to implement this. 

At each iteration, we check if the current element can constitute a valid subarray (i.e. if we add to our subarray will it form a subgroup of **unique** and adjacent block of elements present in array `a`). If it's valid we add the element to the current subarray set and we calculate the `score` of the subarray and then try to expand the subarray window. 

If it's not valid we start to remove elements from the current subarray window (starting with the leftmost element) until we successfully remove the duplicate element from the window (i.e. until window becomes valid again).

## Solution:

```python
class Solution:
    def maximumUniqueSubarray(self, nums: List[int]) -> int:

        # inititalizing hashset that'll store elements in `subarray`  
        subarray = set()

        # initializing `score` (it'll store the sum of elements in current subarray)
        score = 0
        
        # initializing `output` (it'll store maximum score)
        output = 0
        
        # initializing `left` pointer
        left = 0
        
        # iterate through every element position in nums
        for right in range(len(nums)):
            
            # check if the current element is already in `subarray` set
            while nums[right] in subarray:
                
                # remove leftmost integer element from `subarray` set 
                subarray.remove(nums[left])
                
                # decrement score by nums[left]
                score -= nums[left]
                
                # increment `left` pointer 
                # so we can slide window to find the next valid `subarray`
                left += 1
            
            # add current element to subarray set
            subarray.add(nums[right])
            
            # calculate the score of current subarray
            score += nums[right]    
            
            # update output with the maximum score seen so far
            output = max(output, score)
        
        # return output
        return output      
```
