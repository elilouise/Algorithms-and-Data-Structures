# 187. Repeated DNA Sequences

### Difficulty: <font color = orange> Medium </font>

---

The DNA sequence is composed of a series of nucleotides abbreviated as `'A'`, `'C'`, `'G'`, and `'T'`.

For example, `"ACGAATTCCG"` is a DNA sequence.

When studying DNA, it is useful to identify repeated sequences within the DNA.

Given a string s that represents a DNA sequence, return all the `10`<b>-letter-long</b> sequences (substrings) that occur more than once in a DNA molecule. 

You may return the answer in any order.

---


**Example 1:**

**Input**: s = "AAAAACCCCCAAAAACCCCCCAAAAAGGGTTT"

**Output**: $["AAAAACCCCC","CCCCCAAAAA"]$

---

**Example 2:**

**Input**: s = "AAAAAAAAAAAAA"

**Output**: $["AAAAAAAAAA"]$
 
---
**Constraints:**

$1 <= s.length <= 105$

$s[i]$ is either `'A'`, `'C'`, `'G'`, or `'T'`.

## Approach Overview: 

Loop through the character positions in `s`, retrieve the current 10-letter-long substring (sequence) and add it to a set keeping store of all visited susbtrings (sequences). 

If the substring (sequence) is already in the `seenSequences` set, then add it to another set keeping store of all the repeated susbtring in s (we called it `outputSequences`).



## Key Steps:

1. Initializing the set that will store all visited / seen substrings in s

   `seenSequences = set()`


2. Initializing the set that will store all repeated / duplicate substrings in s

   `outputSequences = set()`


3. Loop through the character positions in s (from first character position to 9th before last position) inorder to retrieve and process all 10-letter-long substrings in s without going out of bounds

   `for left in range(len(s) - 9):`


4. Retrieve the current 10-letter-long substring from the current character position

   `currentSequence = s[left: left + 10]`
   
  
5. Check if current substring (sequence) is not in 'seenSequences' set

   `if currentSequence not in seenSequences:`
   

6. If it's not in 'seenSequences' set then add it
   
   `seenSequences.add(currentSequence)`
   
   
7. Check if substring (sequence) is already in 'seenSequences' set

   `else:`


8. If it is, then this means the substring is a duplicate, so add the substring to the output set

   `outputSequences.add(currentSequence)`
   
   
9. Finally, convert the output set to a list and return the results


   `return list(outputSequences)    `

## Solution:

```python
class Solution:
    def findRepeatedDnaSequences(self, s: str) -> List[str]:
        
        # initialize hashset to store all the possible valid sequences in s (DNA molecule)
        seenSequences = set()

        # initialize hashset to store all different repeated sequences in s
        outputSequences = set()

        
        # loop through every character position in s 
        # (omit iterating through the last 9 character positions because 
        # we will go out of bounds)
        for left in range(len(s) - 9):
            
        # retrieve a 10-letter-long sequence (substring) from the current character position in s 
            currentSequence = s[left: left + 10]
            
        # check if the current susbtring (sequence) is not in seenSequences set
            if currentSequence not in seenSequences:
                
                # if not in seenSequences then add it in
                seenSequences.add(currentSequence)
            
            # else check if susbtring (sequence) is in seenSequences set 
            else:
                # add the susbtring (sequence) in output set (outputSequences)
                outputSequences.add(currentSequence)
        
        # convert outputSequences from a set to a list (1d array)
        return list(outputSequences)    
```
