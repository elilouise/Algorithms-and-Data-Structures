# 100. Same Tree

### Difficulty: <font color = green> Easy </font>

---

Given the roots of two binary trees p and q, write a function to check if they are the same or not.

Two binary trees are considered the same if they are structurally identical, and the nodes have the same value.

---


## Approach Overview: 

Traverse through p and q using recursive depth first search. 

Each same positioned nodes we visit, we compare and check they're values. If they're equal we return True, if not we return False.


## Detailed Explanation:

Traverse through each node in p and q (in same order). Each node pair we visit (node p and node q), we check if it has same values. If yes, return True. If no, return False.


## Key Problems:

1. Implementing the base case was tricky, I couldn't get it fully correct. 

So the first thing I needed to consider was what to return when an empty node is reached.

This was relatively easy, if we reach two null nodes (p and q). Then return `True`. (since nodes are the same)

`if p == None and q == None:`

`return True`


What to do when nodes have different values.

`if (p or q is None) or p.val!= q.val:`

`return False`


## Solution:

```python
#Definition for a binary tree node.
#class TreeNode:
#    def __init__(self, val=0, left=None, right=None):
#        self.val = val
#        self.left = left
#        self.right = right
class Solution:
    def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
        
        # check if both p and q are empty
        if p == None and q == None:
            # if yes, 
            # return True (since Trees are the same, they're both empty)
            return True    
        
        # check if p and q have different values or if either p OR q is empty  
        if (p == None or q == None) or (p.val != q.val):
            # if yes 
            # return False (since Trees are not the same, p and q have different values)
            return False
        
        # traverse left subtrees of p and q
        left = self.isSameTree(p.left, q.left)
        
        # traverse right subtrees of p and q 
        right = self.isSameTree(p.right, q.right)
        
        # check if the left & right subtrees of p and q are the same 
        if left and right is True:
            
            # if yes, return True 
            return True
        else:
            
            # else (they're not the same)
            # return False
            return False    
```
