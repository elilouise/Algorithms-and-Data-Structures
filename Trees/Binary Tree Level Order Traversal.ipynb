## Approach Overview:

Perform a **level-order traversal (a Breadth-First Search)** of the binary tree using a queue, processing nodes level by level. For each level, record the node values in a list, then add that list to the final result.


Level-order traversal means visiting all nodes at each level of the binary tree from top to bottom, and from left to right within each level.

## Solution:

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:

        # check if the tree is empty
        if not root:
            # return empty list
            return []

        # initialize our output list
        output = []

        # initializing deque 
        queue = deque([root])
        
        # continue as long as deque is not empty
        while queue:

            level = []  # initialize List to store nodes at the current level

            # loop through all the nodes at the current level of the tree (i.e. all nodes in queue)
            for _ in range(len(queue)):

                # pop the leftmost node in queue 
                node = queue.popleft()

                # add the value of the current node to level 
                level.append(node.val)
                
                # check if the current node has a left child
                if node.left:
                    # add the left child to the queue
                    queue.append(node.left)

                # check if the current node has a right child
                if node.right:
                    # add the right child to the queue
                    queue.append(node.right)
            
            output.append(level)  # Add the current level to the result list
           
        return output
```
