# 226. Invert Binary Tree

### Difficulty: <font color = green> Easy </font>

---
Given the root of a binary tree, invert the tree, and return its root.

---

## Approach Overview:

Recursively transerve through the binary tree (starting with the topmost root node) and swap positions of the children of the current root node.

## Key Steps:

1.  First things first, let's check if the current root Node is valid. A node is considered valid if it's not None. If the node is None, it means we've reached a point in the tree where no further nodes exist.

`if not root:`

`return None` 

Alternative implementation:

`if root == None:`

`return None`


2. Start swapping the left and right nodes with the help of a temporary variable 


3. If we set temporary as `temporary = root.right` then `root.right = root.left` and `root.left = temporary`

   If we set temporary as `temporary = root.left` then `root.left = root.right` and `root.right = temporary`
   
   **Optimization**: *Instead of using a temporary variable to perform the swap the left and right child nodes in one line code without using a temporary variable*
   
   **Efficient Swap:**
   
   `root.right`, `root.left` = `root.left`, `root.right`


4. Repeat the following above steps for the other nodes in the tree using THE POWER OF RECURSION 

`self.invertTree(root.left)` and `self.invertTree(root.right)`


5. finally, return the inverted tree 


   `return root`

## Detailed Explanation:

The challenge was inverting a binary tree. Although the concept was straightforward, implementing it reminded me of the intricacies of working with tree structures compared to arrays and hashmaps. With arrays and hashmaps, we often use iterative approaches, like loops. But with trees, especially binary trees, recursion often feels more natural.

The process of swapping left and right children brought back memories of my A-Level CompSci days, particularly the bubble sort algorithm. This problem, however, was unique because it tapped into the recursive nature of tree structures.

Before each swap, it's crucial to check whether the current node is valid, ensuring we don't operate on non-existent nodes. This was a foundational step at the beginning of our algorithm.

Recursive Call: We further propagated the inversion throughout the tree using:

`self.invertTree(root.left)`

`self.invertTree(root.right)`

By the end, the entire binary tree structure was inverted, achieving the desired outcome.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        
        # The base case
        # Check if node is null
        if not root:
            return None
        
        # Temporary variable to store the left children
        temporary = root.left
        
        # Swap positions of the current children
        
        # Put the right children in the position of the left children
        root.left = root.right
        
        # Put the left children in the position of the right children 
        root.right = temporary
        
        # Repeat the steps above for elements belonging to the right-sided nodes using recursion 
        # Traverse the right sub tree
        self.invertTree(root.right)
        
        # Repeat the steps above for elements belonging to the left-sided nodes using recursion 
        # Traverse the left sub tree
        self.invertTree(root.left)
        
        # Return the inverted binary treeeee
        return root
```
