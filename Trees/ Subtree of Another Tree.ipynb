# 572. Subtree of Another Tree

### Difficulty: <font color = green> Easy </font>

---



## Approach Overview:

Search through the main tree (`root`) and for each node we visit, check if it can start a subtree identical to `subRoot`. 

If we can (if the current node in the main tree (`root`) has a value equal to the root node of `subRoot`), we traverse the subtrees of the current node to check for a match with `subRoot`.

If we can't we keep searching the main tree (`root`) for a match.


## Detailed Explanation:

We do a depth first search through the main tree (`root`) and with each node we visit we check if they can be the start of a subtree identical to `subRoot`. 

We check this by running a depth first search simultaneously through the main tree (`root`) and the `subRoot` tree to check if all the subtrees and nodes in both of these data structures are a match with each other. 

If they are then this means we found a match and we return `True`.

If the nodes do not match, we continue traversing the remaining nodes in the main tree (root) to search for another potential match.



## Key Challenges:

1. Easy to solve conceptually but hard time implementing an algorithmic solution in program code.

## Solutions:

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isSubtree(self, root: Optional[TreeNode], subRoot: Optional[TreeNode]) -> bool:
        
        # Check if the current node matches the start of subRoot
        def sameTree(r, sub):

            # If both nodes are None, weve matched everything in this branch
            if not r and not sub:
                return True
            
            # If only one is None, they dont match (one tree is missing nodes)
            if (not r and sub) or (r and not sub):
                return False
                
            #check if the value of the nodes are the same
            if r.val != sub.val:
                return False
            
            return sameTree(r.left, sub.left) and sameTree(r.right, sub.right)

        # Base case: if root is None, no match is possible
        if not root:
            return False
        
        # Check if the current node matches the start of subRoot
        if sameTree(root, subRoot):
            return True

        # Otherwise, keep looking in the left and right subtrees
        return self.isSubtree(root.left, subRoot) or self.isSubtree(root.right, subRoot)
```
