# 33. Search in Rotated Sorted Array

### Difficulty: <font color = orange> Medium </font>
---

There is an integer array nums sorted in ascending order (with distinct values).

Prior to being passed to your function, nums is possibly rotated at an unknown pivot index `k` (`1 <= k < nums.length`) such that the resulting array is `[nums[k], nums[k+1], ..., nums[n-1], nums[0], nums[1], ..., nums[k-1]]` (`0-indexed`). For example, `[0,1,2,4,5,6,7]` might be rotated at pivot index `3` and become `[4,5,6,7,0,1,2]`.

Given the array nums after the possible rotation and an integer target, return the index of target if it is in nums, or `-1` if it is not in nums.

You must write an algorithm with `O(log n)` runtime complexity.

---

Example 1:

Input: `nums = [4,5,6,7,0,1,2]`, `target = 0`

Output: `4`

---
Example 2:

Input: `nums = [4,5,6,7,0,1,2]`, `target = 3`

Output: `-1`

---
Example 3:

Input: `nums = [1]`, `target = 0`

Output: `-1`
 
---
Constraints:

- `1 <= nums.length <= 5000`
- `-104 <= nums[i] <= 104`
- `All values of nums are unique.`
- `nums is an ascending array that is possibly rotated.`
- `-104 <= target <= 104`

## Approach Overview

Search through nums array for the exact location of target. In each iteration we narrow down search window based on location of target value. 

## Detailed Explanation

Okay so the idea is simple, but the implementation was hard because it was difficult to figure out all the search cases.

What do I mean?

This problem is a classic binary search. We're constantly trying to figure out which portion of the array to ignore and which to narrow our search window to. The difficult part was figuring out the criteria to use to narrow the search range. 

This array is also a rotated array, which means it's split into two sorted parts. One on the left (bigger in size) and one on the right (smaller in size). 

How did we figure out the search criteria? 

Well we used the mid-point, the rightmost, and leftmost element to figure out our current position in the array and the position of the target value itself. Then we narrowed down search window based on those.

In essence, there were only four possibilities.

1) Midpoint is in left section and target is in right section

2) Midpoint is in left section and target is in left section

3) Midpoint is in right section and target is in right section

4) Midpoint is in right section and target is in left section

- If (1) is the case, then we need to move search window to the right section 

- If (2) is the case, then we need to move search window to the left section

- If (3) is the case, then we need to move search window to the right section

- If (4) is the case, then we need to move search window to the left section

<b>Key Note</b>: We're using Midpoint to find target location, each time we compute midpoint, the first thing we check is whether it's now at target location. If not, then we use the pointers (right, left and mid) to firstly find out where target is located, and then we narrow the search window to that location. Rinse and repeat :)

```python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        
        # initializing position of left and right pointer  
        left, right = 0, len(nums) - 1
        
        # continue as long as pointers don't exceed 'nums' bound
        while left <= right:
            
            # calculate mid-point
            mid = (left + right) // 2
            
            # check if current midpoint equals Target
            if nums[mid] == target:

                # return Target position 
                return mid 
            
            # check if midpoint is in left half
            if nums[mid] > nums[right]:

                # Target is located on the left side of midpoint
                if nums[left] <= target <= nums[mid]:
                    
                    # narrow search window to that location by decrementing right pointer
                    right = mid - 1
                
                # Target is located on the right side of midpoint
                else:

                    # narrow search window to that location by incrementing left pointer 
                    left = mid + 1
            
            #Â otherwise midpoint is in left half        
            else:  

                # Target is in the right half
                if nums[mid] <= target <= nums[right]:
                   
                    # narrow search window to that location by incrementing left pointer 
                    left = mid + 1
                
                # otherwise Target is in left half    
                else:
                    
                    # narrow search window to that location by decrementing right pointer
                    right = mid - 1

        # return -1 
        return -1                 
```
