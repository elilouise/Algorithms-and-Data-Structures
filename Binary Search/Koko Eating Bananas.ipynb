# 875. Koko Eating Bananas

### Difficulty: <font color = orange> Medium </font>

---

Koko loves to eat bananas. There are `n` piles of bananas, the `ith` pile has `piles[i]` bananas. The guards have gone and will come back in `h` hours.

Koko can decide her bananas-per-hour eating speed of `k`. Each hour, she chooses some pile of bananas and eats `k` bananas from that pile. If the pile has less than `k` bananas, she eats all of them instead and will not eat any more bananas during this hour.

Koko likes to eat slowly but still wants to finish eating all the bananas before the guards return.

Return the minimum integer `k` such that she can eat all the bananas within `h` hours.

---
Example 1:

Input: piles = [3,6,7,11], h = 8

Output: 4

---
Example 2:

Input: piles = [30,11,23,4,20], h = 5

Output: 30

---
Example 3:

Input: piles = [30,11,23,4,20], h = 6

Output: 23
 
---
Constraints:

- 1 <= piles.length <= 104
- piles.length <= h <= 109
- 1 <= piles[i] <= 109

## Approach Overview:

Test if koko can eat bananas in time against a range of 'eating speed' solution set. We perform binary search on the solution set (koko's eating speed) inorder to lookup and test potentially valid 'eating speed' solutions more efficiently (faster).

- If the eating speed is fast enough (eating time is less than or equal to h), then we save that value, and consider an even smaller valid eating speed until we find the most minimum valid eating speed from the solution set.

We then return that most minimum valid eating time :)

# Solution:

```python
class Solution:
    def minEatingSpeed(self, piles: List[int], h: int) -> int:
        
        # minimum possible eating speed (k_min)
        left = 1

        # maximum possible eating speed (k_max)
        right = max(piles)
        
        # initialize output to +ve infinity
        output = float("inf")

        # continue as long as we've 
        while left <= right:
            
            # calculate the eating speed of koko (bananas-per-hour)
            k = (left + right) // 2
            
            # number of hours it takes koko to eat all the piles of bananas
            hours = 0
            
            # iterate through every bananas piles
            for pile in piles:
                
  # calculate the number of hours it will take koko to eat all the banana piles
                hours = hours + math.ceil(pile / k)
            
            # check if eating time (hours) is larger than h
            if hours > h:
                
    # if eating time is too long, then consider the next successive larger eating speed 
    # we need to increase the eating speed because eating time is too long (too slow)
                left = k + 1
            
            # check if eating time is less than or equal to h
            else: 
            
            # if eating time is valid, then let's try look for an even smaller valid one 
            # we'll do that by looking at the smaller adjacent eating time (located to the left of the current k value / eating time)
                right = k - 1

            # update the minimum valid eating time 
                output = min(output, k)
       
        # return minimum valid eating time    
        return output        
```
